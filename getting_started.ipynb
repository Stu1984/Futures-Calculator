```python
initial_deposit = 200

days_compound_only = 365

frequency_of_compounds = 14

ongoing_deposit_amount = 200

reward_rate = 0.005

balance = initial_deposit

for index, i in enumerate(range(0,days_compound_only)):
    if i % frequency_of_compounds == 0:
        daily_reward = balance * reward_rate
        total_rewards = frequency_of_compounds * daily_reward
        balance = balance + ongoing_deposit_amount + total_rewards
        print(index, round(total_rewards, 2), round(balance, 2))
    else:
        print(index, round(total_rewards, 2), round(balance, 2))
        continue


```
